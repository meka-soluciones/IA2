#Mismo ejercicio pero con Sklearn

#1. Librerias

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.datasets import make_regression
```

#2. Data Set

```python
x, y = make_regression(n_samples=500, n_features=1, noise=20)
```

#3. Crear el modelo

```python
modelo=LinearRegression(fit_intercept=True)
```

#4. Entrenar el modelo

```python
modelo.fit(x,y)
```

#5. Graficar

```python
plt.figure(figsize=(12,8))
plt.scatter(x,y, c=y)
plt.plot(x,modelo.predict(x),color='red')
plt.title('Regresion Gradiente Descendiente')
plt.xlabel('x: Caracteristica de entrada')
plt.ylabel('y: Target o valor real')
plt.show()
```

#6. Ecuacion del modelo

```python
theta0=modelo.intercept_
theta1=modelo.coef_[0]
print(f'h(x)={theta0:.2f}+{theta1:.2f}x')
```

#7. Entrenamiento con split

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
modelo.fit(x_train,y_train)
```

```python
plt.plot(x_test,modelo.predict(x_test),color='red')
plt.scatter(x_test,y_test)
plt.title('Regresion Gradiente Descendiente con Split')
plt.xlabel('x: Caracteristica de entrada')
plt.ylabel('y: Target o valor real')
plt.show()
```

#8. Conclusiones

1. Aprendi que al Realizar el entrenamiento con Split a una particion del data Set; se puede obtener una mejor prediccion.
2. El comando .predict realiza la prediccion del modelo con la particion del data set que se designe, en este caso x_test y y_test. De otra manera solo podria comparar el modelo con si mismo.
