#Importamos el data set de clasificacion y realizamos la grafica

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_classification
x, y = make_classification(
    n_samples=100,
    n_features=2,
    n_redundant=0,
    n_informative=2,
    random_state=3,
    n_clusters_per_class=1,
)
plt.scatter(x[:, 0], x[:, 1], c=y, cmap=plt.cm.Spectral)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.33, random_state=42)
print(x[0:10])
x.shape
```

#Establecemos la arquitectura de la red Neuronal

```python
import tensorflow as tf
modelo=tf.keras.Sequential([tf.keras.layers.Dense(4, activation='relu',input_shape=(2,)),tf.keras.layers.Dense(6, activation='relu'),tf.keras.layers.Dense(2,activation='softmax')])
```

*tf.keras.layers establece el numero de neuronas en la capa.
       
*Si se establece una salida Softmax (promedio) se necesitan de 2 neuronas activas.

*input_shape establece el numero de neuronas de entrada a la capa.

#2. Compilar el modelo

```python
modelo.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

#3. Entrenar el modelo

```python
modelo.fit(X_train, Y_train, epochs=50)
```
